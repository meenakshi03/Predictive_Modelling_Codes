```python
import re
```

```python
# Creating the corpus along with structuring data in a readable form
corpus = []
for title_abs in title:
    title_abs = str(title_abs)
    title_abs = title_abs.lower()
    title_abs = re.sub(r"that's","that is",title_abs)
    title_abs = re.sub(r"there's","there is",title_abs)
    title_abs = re.sub(r"what's","what is",title_abs)
    title_abs = re.sub(r"where's","where is",title_abs)
    title_abs = re.sub(r"it's","it is",title_abs)
    title_abs = re.sub(r"who's","who is",title_abs)
    title_abs = re.sub(r"i'm","i am",title_abs)
    title_abs = re.sub(r"she's","she is",title_abs)
    title_abs = re.sub(r"he's","he is",title_abs)
    title_abs = re.sub(r"they're","they are",title_abs)
    title_abs = re.sub(r"who're","who are",title_abs)
    title_abs = re.sub(r"ain't","am not",title_abs)
    title_abs = re.sub(r"wouldn't","would not",title_abs)
    title_abs = re.sub(r"shouldn't","should not",title_abs)
    title_abs = re.sub(r"can't","can not",title_abs)
    title_abs = re.sub(r"couldn't","could not",title_abs)
    title_abs = re.sub(r"won't","will not",title_abs)
    title_abs = re.sub(r"\W"," ",title_abs)
    title_abs = re.sub(r"\d"," ",title_abs)
    title_abs = re.sub(r"\s+[a-z]\s+"," ",title_abs)
    title_abs = re.sub(r"\s+[a-z]$"," ",title_abs)
    title_abs = re.sub(r"^[a-z]\s+"," ",title_abs)
    title_abs = re.sub(r"\s+"," ",title_abs)
    title_abs = re.sub(r'^br$', ' ', title_abs)
    title_abs = re.sub(r'\W', ' ',title_abs)
    title_abs = title_abs.rstrip() #to remove trailing whitespaces at the end of the string
    title_abs = title_abs.lstrip() #to remove leading whitespaces at the start of the string
    corpus.append(title_abs) 
```

```python
#string of the corpus 
str_representation = ' '.join(corpus)

#list of words of the corpus
list_words=str_representation.split(" ")
print('Number of words in the corpus is : ',len(list_words))

#unique count of words
print('Number of unique words in the corpus are : ',len(set(list_words)))
```

```python
## Removing stopwords( here i am doing it before lemmetization as latter converts words like 'has' to 'ha' and so if stop words 
## removal done after lemmetization then 'ha' wont be captured)
import nltk
from nltk.corpus import stopwords

for i in range(len(corpus)):
    words = nltk.word_tokenize(corpus[i])
    words = [word for word in words if word not in stopwords.words('english')]
    corpus[i] = ' '.join(words)
```

### Predictiong probability for tweeking the cutoff and finding the ideal cut_off

```python
test_train_predicted_prob = pd.DataFrame(train_rf_tree.predict_proba(rf_test_X))[[1]]
test_train_predicted_prob.columns = ['prob']
actual=rf_test_y.reset_index()
actual.drop('index',axis=1,inplace=True)
# making a DataFrame with actual and prob columns
test_train_predict = pd.concat([actual, test_train_predicted_prob], axis=1)
test_train_predict.columns = ['actual','prob']
test_train_predict.head()
```

```python
test_train_predict.shape
```

```python
#Intution behind ROC curve - confusion matrix for each different cut-offs shows trade off in sensivity and specificity
test_roc_like_df=pd.DataFrame()
test_temp=test_train_predict.copy()
```

```python
for cut_off in np.linspace(0,1,50):
    test_temp['predicted']=test_temp.prob.apply(lambda x:0 if x < cut_off else 1)
    test_temp['tp']=test_temp.apply(lambda x: 1 if x['actual']==1 and x['predicted']==1 else 0,axis=1)
    test_temp['fp']=test_temp.apply(lambda x: 1 if x['actual']==0 and x['predicted']==1 else 0,axis=1)
    test_temp['tn']=test_temp.apply(lambda x: 1 if x['actual']==0 and x['predicted']==0 else 0,axis=1)
    test_temp['fn']=test_temp.apply(lambda x: 1 if x['actual']==1 and x['predicted']==0 else 0,axis=1)
    sensitivity=test_temp['tp'].sum()/(test_temp['tp'].sum()+test_temp['fn'].sum())
    specificity=test_temp['tn'].sum()/(test_temp['tn'].sum()+test_temp['fp'].sum())
    test_roc_like_table=pd.DataFrame([cut_off,sensitivity,specificity]).T
    test_roc_like_table.columns=['cut_off','sensitivity','specificity']
    test_roc_like_df=pd.concat([test_roc_like_df,test_roc_like_table],axis=0)
    
test_temp.sum()
plt.subplots(figsize=(10,4))
plt.scatter(test_roc_like_df['cut_off'],test_roc_like_df['sensitivity'],marker='*',label='Sensitivity',)
plt.scatter(test_roc_like_df['cut_off'],test_roc_like_df['specificity'],marker='o',label='Specificity')
#plt.scatter(test_roc_like_df['cut_off'],1-test_roc_like_df['specificity'],marker='*',label='FPR')
plt.title('For each cutoff, pair of sensitivity and FPR is plotted for ROC')
plt.legend()
```

```python
##Finding ideal cut off
test_roc_like_df['total']=test_roc_like_df.sensitivity+test_roc_like_df.specificity
test_roc_like_df[test_roc_like_df.total==test_roc_like_df.total.max()]
```
